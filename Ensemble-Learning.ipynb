```python
import pandas as pd
import numpy as np
from sklearn.ensemble import BaggingClassifier, BaggingRegressor
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn.ensemble import ExtraTreesClassifier, ExtraTreesRegressor
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from xgboost import XGBClassifier, XGBRegressor
from sklearn.ensemble import VotingClassifier, VotingRegressor
from sklearn.ensemble import StackingClassifier, StackingRegressor
from sklearn.neighbors import KNeighborsClassifier , KNeighborsRegressor
from sklearn.linear_model import LinearRegression , LogisticRegression
from mlxtend.classifier import StackingClassifier as stc
from mlxtend.regressor import StackingRegressor as strg
from sklearn.model_selection import train_test_split,GridSearchCV,KFold
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
data = pd.read_csv("C:\\Thanusha\\Python\\ML master class\\Ensemble\\Bank.csv")
data.head()
```

```python
x = data.iloc[:,:7].values
y = data.iloc[:,7].values
```

```python
x_tain,x_test,y_train,y_test = train_test_split (x,y, test_size=0.2 , random_state=0)
```

```python
## classification models ,   cm = classification model
```

```python
cm1 = BaggingClassifier()
cm1.fit(x_tain,y_train)
```

```python
cm2 = RandomForestClassifier()
cm2.fit(x_tain,y_train)
```

```python
cm3 = ExtraTreesClassifier()
cm3.fit(x_tain,y_train)
```

```python
cm4 = AdaBoostClassifier()
cm4.fit(x_tain,y_train)
```

```python
cm5 = GradientBoostingClassifier()
cm5.fit(x_tain,y_train)
```

```python
cm6 = XGBClassifier()
cm6.fit(x_tain,y_train)
```

```python
accuracy_score(y_test, cm1.predict(x_test))
```

```python
## Voting Classifier 
```

```python
# hard voting
```

```python
m1 = KNeighborsClassifier(n_neighbors=5)
m2 = LogisticRegression()
m3 = RandomForestClassifier()
```

```python
vc = VotingClassifier(estimators= [("knn", m1),("lgr" , m2),("rf", m3)], voting="hard")
```

```python
## "knn," "lgr" are nicknames. 
```

```python
vc.fit(x_tain,y_train)
```

```python
y_pred = vc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
## soft voting
```

```python
vcs = VotingClassifier(estimators= [("knn", m1),("lgr" , m2),("rf", m3)], voting="soft")
```

```python
vcs.fit(x_tain,y_train)
```

```python
y_pred = vcs.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
## Stacking classifier - SkLearn
```

```python
# hererogeneous models
T1 = ("lgr", LogisticRegression())
T2 = ("knn", KNeighborsClassifier(n_neighbors=5))
```

```python
# meta model
metamodel = RandomForestClassifier()
```

```python
stcl = StackingClassifier(estimators=[T1,T2], final_estimator=metamodel)
```

```python
stcl.fit(x_tain,y_train)
```

```python
y_pred = stcl.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
## Stacking classifier - MLXtend
```

```python
bm1 = LogisticRegression()
bm2 = KNeighborsClassifier(n_neighbors=5)
mm = RandomForestClassifier()
```

```python
stcl = stc (classifiers=[bm1,bm2], meta_classifier=mm)
```

```python
stcl.fit(x_tain,y_train)
```

```python
y_pred = stcl.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
## Stacking Regressor - MLXten
```

```python
bm1 = LinearRegression()
bm2 = KNeighborsRegressor(n_neighbors=5)
mm = RandomForestRegressor()
```

```python
model = strg(regressors = [bm1,bm2], meta_regressor=mm)
```

```python
model.fit(x_tain,y_train)
```

```python
y_pred = model.predict(x_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```
